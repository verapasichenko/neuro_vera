# Краткие сведения

Проект представляет из себя реализацию стойкого шифрования стегоизображения с последующим извлечением информации с помощью нейронной сети, модель построена на основе ResNet и показаывает высокую точность.

# Настройка


1.  Настройте virtualenv с помощью python
2.  Запустите pip install -r requirements.txt, чтобы загрузить библиотеки

codder.py - используйте для шифрования и дешифрования изображения

net.ipynb - модель

net.py - класс сети для использования ее в codder.py

metrics.py - измерение параметров встраивания при атаке на изображение

prepairing.py - функции которые вам понадобятся для подготовки датасета

---



*Если вы хотите использовать заранее обученную модель выполните следующие шаги:*

model = ClassificatorNet(*args, **kwargs)

model.load_state_dict(torch.load(PATH))

model.eval()
